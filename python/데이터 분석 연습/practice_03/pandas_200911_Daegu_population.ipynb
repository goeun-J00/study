## 1. 인구를 10살 단위로 나누고 70세 이상은 하나로 묶는다  

```python
import pandas as pd
import numpy as np
df_Jun18 = pd.read_csv('data/2018.06.30.기준 대구광역시 주민등록인구 현황_UTF8.csv')
df_Dec18 = pd.read_csv('data/2018.12.31.기준 대구광역시 주민등록인구 현황_UTF8.csv')
df_Jun19 = pd.read_csv('data/2019.06.30.기준 대구광역시 주민등록인구 현황_UTF8.csv', thousands = ',')
```

```python
df_Jun19 = df_Jun19.rename(columns = {"행정구역" : "구분"})
df_Jun19.head()
```

```python
def drop_col(df, check_list=[]):
    drop_list = []
    for chk in check_list:
        for col in df.columns :
            if  chk in col:
                drop_list.append(col)
    drop_list = list(set(drop_list)) #  중복 제거를 위해 set을 사용
    return drop_list
```

```python
def age_groub(df):
    df = df.drop(drop_list, axis=1)
    df['연령'] = (pd.DataFrame(range(0,101))/10).astype(int)*10
    df['연령'] = np.where(df['연령'] >= 70, 70, df.연령)
    df = df.groupby('연령')
    df = df.sum()
    return df
```

```python
check_list = ['구분', '총계', ' 남', ' 여']
drop_list = drop_col(df_Jun18, check_list)
df1 = age_groub(df_Jun18)
df2 = age_groub(df_Dec18)
df3 = age_groub(df_Jun19)
df1
```

## 2. 3개의 기간동안 인구가 계속 증가하는 지역은 어디인가? 

```python
df1 = df1.T
df1['총합'] = df1.sum(axis='columns').astype(int)
df1
```

```python
df2 = df2.T
df2['총합'] = df2.sum(axis='columns').astype(int)
df2
```

```python
df3 = df3.T
df3['총합'] = df3.sum(axis='columns').astype(int)
df3
```

```python
pop1 = (df2['총합'] - df1['총합']) > 0
pop2 = (df3['총합'] - df2['총합']) > 0
df3['인구증가'] = pop1 & pop2
df3
```

## 3. 남성이 여성보다 많이 거주하는 곳, 인구가 20만 이상인 곳

```python
check_list = ['구분', '계', '여']
drop_list = drop_col(df_Jun18, check_list)
df1_male = age_groub(df_Jun18)
df2_male = age_groub(df_Dec18)
df3_male = age_groub(df_Jun19)
df1_male
```

```python
def df_op(df0,df1):
    df0.index = ['중구','동구','서구','남구','북구','수성구','달서구','달성군']
    #print(df0)
    df1.index = ['중구','동구','서구','남구','북구','수성구','달서구','달성군']
    #print(df1)
    df_pluse = df0 + df1
    df_minus = df0 - df1
    return df_pluse, df_minus
```
